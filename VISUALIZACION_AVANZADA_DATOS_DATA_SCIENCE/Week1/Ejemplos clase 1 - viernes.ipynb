<h2>Visualización de datos con Matplotlib</h2>

[Matplotlib](https://matplotlib.org/stable/index.html) es una librería en python para crear visualizaciones estáticas e interactivas.

<img src="./images/figure.png" width="55%">

En matplotlib existen dos tipos de formas de definir y construir visualizaciones:
- **Explicita:** *OOS* - *object-oriented style*. Se crean la figura y los ejes, sobre los cuales se empiezan a llamar los métodos para graficar.
- **Implicita:** Se utilizan funciones de `pyplot` para graficar. Este se encarga de gestionar el *OOS* para generar las visualizaciones.

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Ejemplo de la forma **explicita**, donde definimos un `subplot` el cual retorna una figura `fig` y unos ejes `ax`. Note que logramos graficar haciendo uso de los llamados a los ejes `ax`.

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(8, 4), layout='constrained')
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label') 
ax.set_title("Simple Plot using Object-oriented style")
ax.legend();
```

En la forma **implicita** no definimos una figura o ejes. Lo que hacemos es llamar funciones de `pyplot`, y este se va encargando de construir nuestra gráfica.

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(8, 4), layout='constrained')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot using pyplot-style")
plt.legend();
```

<h3>Propiedades de las figuras</h3>

Para definir una figura de la forma implícita lo hacemos mediante el llamado a la función [`plt.figure()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure). En la cual podemos definir diferentes parámetros:
- `figsize`: tupla `(width, height)` para definir el tamaño de la figura.
- `facecolor`: color del fondo de la figura.
- `constrained_layout`: Ajustar espaciado de los elementos de la figura.

```python
plt.figure(figsize=(5,3))
title_prop = {'fontsize': 16,
              'color': "blue",
              'horizontalalignment': "center"}
plt.title("Ejemplo de titulo de un gráfico", fontdict=title_prop)
plt.xlabel("Etiqueta en el eje x", loc="right", rotation=20)
plt.ylabel("Etiqueta en el eje y")
plt.grid();
```

<h3>Tipos de gráficos</h3>

[Matplotlib](https://matplotlib.org/stable/index.html) ofrece un gran número de [representaciones gráficas](https://matplotlib.org/stable/plot_types/index.html#plot-types). Las cuales mediante la llamada a la función de cada gráfico nos retorna (mediate el uso implícito) el gráfico.

- `plt.plot(**args)`: Nos permite graficar lineas, relaciones en 2D
- `plt.scatter(**args)`: Graficos de dispersión
- `plt.bar(**args)`: Graficos de barras
- `plt.hist(**args)`: Nos permite graficar distribuciones
- `plt.boxplot(**args)`: Para representar diagramas de cajas
- `plt.violinplot(**args)`: Para representar diagramas de violín
- `plt.pie(**args)`: Para representar proporciones

Exploremos `plt.plot(**args)`

```python
y_1 = np.linspace(0,10, 100)
x_1 = range(len(y_1))

plt.figure(figsize=(6,4))
plt.title("Ejemplo de un plot")
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x_1, y_1, color="red", ls=":")

plt.grid();
```

```python
#Generar datos aleatorios
import pandas as pd

np.random.seed(4)
metros_cuadrados = np.random.randint(50, 150, size=10)
habitaciones = np.random.randint(1, 5, size=10)
precio = metros_cuadrados * np.random.randint(800, 1500, size=10)

df = pd.DataFrame({
    "metro_cuadrado": metros_cuadrados,
    "numero_habitaciones": habitaciones,
    "precio": precio
})

df
```

```python
plt.figure(figsize=(8,5))
plt.title("Relación de metro cuadrado vs precio", fontsize=16)
plt.scatter(df["metro_cuadrado"], df["precio"])
plt.xlabel("metros cuadrados")
plt.ylabel("precio")
plt.grid();
```

```python
df.numero_habitaciones.values
```

```python
plt.figure(figsize=(8,5))
plt.title("Agregando una tercera relacion r=habitaciones", fontsize=16)
plt.scatter(df["metro_cuadrado"], df["precio"], s=df["numero_habitaciones"]*100)
plt.xlabel("metros cuadrados")
plt.ylabel("precio")
plt.grid();
```

Podemos agregar [anotaciones](https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Annotation) sobre el gráfico:

```python
plt.figure(figsize=(8,5))
plt.title("Agregando una tercera relacion r=habitaciones", fontsize=16)
plt.scatter(df["metro_cuadrado"],
            df["precio"],
            s=df["numero_habitaciones"]*80,
           color="#AFE394")
for x, y, text in zip(df["metro_cuadrado"], df["precio"], df["numero_habitaciones"]):
    plt.annotate(text,(x,y))
plt.xlabel("metros cuadrados")
plt.ylabel("precio")
plt.grid();
```

[markers](https://matplotlib.org/stable/api/_as_gen/matplotlib.markers.MarkerStyle.html#matplotlib.markers.MarkerStyle)

```python
plt.figure(figsize=(8,5))
plt.title("Agregando una tercera relacion r=habitaciones", fontsize=16)
plt.scatter(df["metro_cuadrado"],
            df["precio"],
            s=df["numero_habitaciones"]*80,
           color="#AFE394",
           marker="v")
for x, y, text in zip(df["metro_cuadrado"], df["precio"], df["numero_habitaciones"]):
    plt.annotate(text,(x,y))
plt.xlabel("metros cuadrados")
plt.ylabel("precio")
plt.grid();
```
